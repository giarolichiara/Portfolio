```python
import csv

#dizionario punteggi-posizioni
scores = {
    "1": 10,
    "2": 8,
    "3": 6,
    "4": 5,
    "5": 4,
    "6": 3,
    "7": 2,
    "8": 1
}

#funzione per performance individuali dei piloti
def individual_performance():
    total_scores = 0
    wins = 0
    podiums = 0
    driver_names = set()
    
    with open("formula1_data.csv") as csv_file:
        csv_reader = csv.reader(csv_file)

        for row in csv_reader:
            driver = row[0].strip()
            driver_names.add(driver.lower())

    while True:
        name_driver = input("Inserisci il nome di un pilota: ")
        if name_driver.lower() not in driver_names:
            print("Il nome del pilota scelto non è presente nel file CSV. Riprova!")
        else:
            break

    with open("formula1_data.csv") as csv_file:
        csv_reader = csv.reader(csv_file)
        
        for row in csv_reader:
            driver = row[0].strip()
            position = row[4].strip()

            if driver.lower() == name_driver.lower():
                if position.isdigit():
                    position_int = int(position)

                    """Punteggio"""
                    total_scores += scores.get(position, 0)

                    """Vittoria"""
                    if position_int == 1:
                        wins += 1

                    """Podio"""
                    if position_int <= 3:
                        podiums += 1

    return [name_driver, total_scores, wins, podiums]


#funzione per classifica dei piloti
def driver_ranking():
    driver_ranking = {}

    with open("formula1_data.csv") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)

        for row in csv_reader:
            name_driver = row[0].strip()
            position = row[4].strip()

        
            if position in scores:
                point = scores[position]
            else:
                point = 0

            if name_driver in driver_ranking:
                driver_ranking[name_driver] += point
            else:
                driver_ranking[name_driver] = point

    #trasformo in lista per l'ordine descrescente dei punteggi
    list_ranking = list(driver_ranking.items())
    list_ranking.sort(key=lambda x:x[1], reverse=True)

    #ritrasformo lista in dizionario ma ordinato
    d_ranking_dict = dict(list_ranking)

    with open("Drivers_Standings_2008.txt", "w") as new_file:
        new_file.write("Driver Standings 2008 Formula 1\n")
        for pilot, point in d_ranking_dict.items():
            new_file.write(f"{pilot}: {point}\n")
            
    print("Classifica dei piloti salvata in un nuovo file")


#funzione per classifica dei team
def team_ranking():
    t_ranking = {}
    
    with open("formula1_data.csv") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)

        for row in csv_reader:
            team = row[1].strip()
            position = row[4].strip()

            if position in scores:
                point = scores[position]
            else:
                point = 0

            if team in t_ranking:
                t_ranking[team] += point
            else:
                t_ranking[team] = point

    #lista classifica in ordine decrescente
    team_list = list(t_ranking.items())
    team_list.sort(key=lambda x: x[1], reverse=True)

    #ritrasformo in dizionario
    t_ranking_dict = dict(team_list)

    with open("Teams_Standings_2008.txt", "w") as new_file:
        new_file.write("Teams Standings 2008 Formula 1\n")
        for team, point in t_ranking_dict.items():
            new_file.write(f"{team}: {point}\n")
            
    print("Classifica dei team salvata in un nuovo file")


result = individual_performance()
print(f"{result[0]} ha totalizzato un totale di {result[1]} punti, con {result[2]} vittorie e {result[3]} podi totali ")

save_driver_file = input("Vuoi salvare la classifica dei PILOTI in un file? (si/no): ")
if save_driver_file == "si":
    driver_ranking()
    
save_team_file = input("Vuoi salvare la classifica dei TEAM in un file? (si/no): ")
if save_team_file == "si":
    team_ranking()
```
